```python
df_avg.info()
```

```python
Nondf_avg = df_avg.dropna()
Nondf_avg

x = Nondf_avg.drop(columns='평균임금')
y = Nondf_avg['평균임금']
```

```python
Nandf_avg=df_avg.drop(Nondf_avg.index)
Nandf_avg

x2  = Nandf_avg.drop(columns='평균임금')
y2  = Nandf_avg['평균임금']
```

```python
model= XGBRegressor()
model.fit(x,y)
```

```python
Nandf_avg['평균임금'] = model.predict(x2)
Nandf_avg
```

```python
df_pred = pd.concat([Nondf_avg,Nandf_avg]).sort_index()
df_pred = df_pred.astype({'직종': int, '평균임금':int})
df_pred
```
